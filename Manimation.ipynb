```python
%%manim -qm -v WARNING SquareToCircle

class SquareToCircle(Scene):
   def construct(self):
      square = Square()
      circle = Circle()
      circle.set_fill(PINK, opacity=0.5)
      self.play(Create(square))
      self.play(Transform(square, circle))
      self.wait()
```

```python
%%manim -qm -v WARNING BubbleSortAnimation

from manim import *

class BubbleSortAnimation(Scene):
    def construct(self):
        values = [7, 2, 9, 1, 5]
        rects = []

        # Create rectangles to represent the values
        for i in range(len(values)):
            rect = Rectangle(height=values[i], width=0.5, fill_opacity=1)
            rect.next_to(rects[-1], RIGHT, buff=0.5) if rects else rect.shift(2 * LEFT)
            rects.append(rect)

        self.play(*[Create(rect) for rect in rects])
        self.wait()

        # Bubble sort algorithm
        for i in range(len(values)-1):
            for j in range(len(values)-1-i):
                if values[j] > values[j+1]:
                    values[j], values[j+1] = values[j+1], values[j]
                    self.play(Swap(rects[j], rects[j+1], run_time=0.5))

        self.wait(2)

        self.play(*[FadeOut(rect) for rect in rects])

BubbleSortAnimation().render()
```

```python
%%manim -qm -v WARNING StartingAnimation

from manim import *

class StartingAnimation(Scene):
    def construct(self):
        # Create the objects
        objects = VGroup(
            Circle(radius=0.5, fill_opacity=1, color=YELLOW),
            Square(side_length=1, fill_opacity=1, color=BLUE),
            RegularPolygon(n=3, fill_opacity=1, color=GREEN)
        )

        # Position the objects
        objects.arrange(RIGHT, buff=1)
        objects.scale(1.5)
        objects.to_edge(UP)

        # Create the channel name
        channel_name = Text("Twist Nemo", font_size=48, color=WHITE)
        channel_name.next_to(objects, DOWN)

        # Animation sequence
        self.play(GrowFromCenter(objects), run_time=2)
        self.wait(1)
        self.play(Transform(objects, channel_name), run_time=1)
        self.wait(2)
        self.play(FadeOut(objects), run_time=1)

# Render the animation
scene = StartingAnimation()
scene.render()
```

```python
%%manim -qm -v WARNING LoopAnimationJupyter

from manim import *

class LoopAnimationJupyter(Scene):
    def construct(self):
        # Create the loop counter
        counter = Integer(0, font_size=48)
        counter.to_edge(UP)

        self.play(Write(counter))

        # Create the loop animation
        loop_text = Text("Looping...", font_size=32, color=GREEN)
        loop_text.move_to(ORIGIN)

        self.play(Transform(counter, loop_text))
        self.wait(1)

        # Perform the loop iterations
        for i in range(1, 6):
            iteration_text = Text(f"Iteration {i}", font_size=32, color=YELLOW)
            iteration_text.move_to(ORIGIN)

            self.play(Transform(counter, iteration_text))
            self.wait(1)

        # Reset the counter
        reset_text = Text("Reset", font_size=32, color=RED)
        reset_text.move_to(ORIGIN)

        self.play(Transform(counter, reset_text))
        self.wait(1)

        # Explain the loop process
        explanation = Text("The loop repeats", font_size=32, color=BLUE)
        explanation.next_to(counter, DOWN)

        arrow = Arrow(start=ORIGIN, end=counter.get_center() + DOWN, buff=0.1)

        self.play(Write(explanation), Create(arrow))
        self.wait(2)

        # Fade out and clean up
        self.play(FadeOut(explanation), FadeOut(arrow))
        self.wait(1)
        self.play(FadeOut(counter))
        self.wait(1)
```
